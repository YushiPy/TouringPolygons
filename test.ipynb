## Setup

```python
import traceback
from typing import Iterable, Sequence
from vector2 import Vector2
from polygon2 import Polygon2

from problem2_draw import Drawing

type CanVector = Sequence[float] | Vector2
type CanPolygon = Iterable[CanVector]
type Test = tuple[CanVector, CanVector, Iterable[CanPolygon], Iterable[CanPolygon]]

def do_test(start: CanVector, end: CanVector, polygons: Iterable[CanPolygon], fences: Iterable[CanPolygon]) -> None:
	"""
	Test the drawing functionality with given start and end points, polygons, and expected output.
	"""

	_start = Vector2(*start)
	_end = Vector2(*end)

	_polygons = [Polygon2(polygon) for polygon in polygons]
	_fences = [Polygon2(fence) for fence in fences]

	drawing = Drawing(_start, _end, _polygons, _fences)
	
	try: 
		drawing.shortest_path()
	except ValueError:
		traceback.print_exc()

	drawing.draw()

def parse_test(values: list[list[tuple[float, float]]]) -> Test:
	"""
	Parse a list of values into a Test tuple.
	"""

	ones = [i for i in values if len(i) == 1]
	values = [i for i in values if len(i) > 1]

	start = ones[0][0]
	end = ones[1][0]
	
	polygons = values[1::2]
	fences = values[::2]

	return start, end, polygons, fences
```

## Tests

```python
x = [
	[(0.0, 4.0)],
	[(-6.0, 5.0), (-6.0, 1.0), (-1.0, 1.0), (-6.0, -0.0), (-6.0, -3.0), (6.0, -3.0), (6.0, -0.0), (1.0, 1.0), (6.0, 1.0), (6.0, 5.0)],
	[(-3.0, -0.0), (3.0, -0.0), (0.0, -1.0)],
	[(4.0, -2.0)],
	[(-7.0, 2.0), (7.0, 2.0), (7.0, -4.0), (-7.0, -4.0)],
]

do_test(*parse_test(x))
```
